```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
x = np.linspace(1,10,100)
```

```python
sin_x = np.sin(x)
```

```python
cos_x = np.cos(x)
```

```python
tan_x = np.tan(x)
```

```python
plt.plot(x,sin_x, 'b--')
plt.plot(x,cos_x, 'r--')
#plt.plot(x,tan_x, 'g--')
plt.show()
```

```python
sns.set_style('ticks')
plt.plot(x,sin_x, 'b--')
plt.plot(x,cos_x, 'r--')
#plt.plot(x,tan_x, 'g--')
plt.show()
```
