```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = np.genfromtxt('iris.data', delimiter=',', usecols=(0,1,2,3))
data
```

```python
plt.plot(data[:50,0],c='Red', ls=':',marker='s', ms=8)
```

```python
plt.plot(data[50:100,0],c='Blue', ls='--',marker='s', ms=8)
```

```python
plt.plot(data[:50,0],c='Red', ls=':',marker='s', ms=8, label='Comprimento Spala Iris-Setosa')
plt.plot(data[50:100,0],c='Blue', ls=':',marker='o', ms=8, label='Comprimento Spala Iris-Versicolor')
plt.plot(data[100:150,0],c='Green', ls=':',marker='^', ms=8, label='Comprimento Spala Iris-Virginica')
plt.legend()
plt.show()
```
