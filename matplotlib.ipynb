```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.linspace(1,10, 100)
```

```python
plt.plot(x, np.sin(x))
```

```python
plt.plot(x, np.cos(x))
```

```python
plt.plot(x, np.tan(x))
```

```python
plt.plot(x, np.sin(x), 'r')
```

```python
plt.plot(x, np.sin(x), '--')
```

```python
plt.plot(x, np.sin(x), 'r--')
```
